CPI Analysis per country ( 2021 onwards)

```python
file_path = Path("Resources/CPI rates.csv")
CPI_data = pd.read_csv(file_path)
CPI_data.head()
```

```python
CPI_data = CPI_data.set_index(["Series Name", "Country Name","2021[YR2021]"])
columns_to_keep = [col for col in CPI_data.columns if col.isdigit() and int(col) >= 2019]
filtered_dataframe = CPI_data[columns_to_keep]
filtered_dataframe
```

```python
continent_data_path = "Resources/owid-covid-data.csv"
continent_data = pd.read_csv(continent_data_path)
combined_df = pd.merge(continent_data,CPI_data,on="COUNTRY")
combined_df
```

```python


plt.figure(figsize=(10, 6))


for index, row in filtered_dataframe.iterrows():
    plt.plot(filtered_dataframe.columns, row, marker='o', label=f'{index[0]}')


plt.title('CPI Trend (2020 per continent)')
plt.xlabel('Continent')
plt.ylabel('2020')
plt.legend(title="Country", loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)


plt.grid(True)
plt.scatter()
plt.show()
```

Vaccinations Analysis 

```python
file_path = Path("Resources/vaccination-data.csv")
Vaccine_data = pd.read_csv(file_path)
Vaccine_data.head()
```

```python

data =Vaccine_data
columns_to_exclude = ["]
df_filtered = data.drop(columns =columns_to_exclude)
df_filtered
```

```python
Countries and Continent merges 
```

```python
First Dose 
```

```python
df_grouped = df_filtered.groupby('WHO_REGION').sum()

# Plotting the country vs total vaccinations
plt.figure(figsize=(10, 6))
df_grouped["PERSONS_VACCINATED_1PLUS_DOSE_PER100"].plot(kind='bar')

# Adding title and labels
plt.title('Persons vaccinated by first dose vs WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Persons vaccinated by first dose(per 100)')

# Displaying the plot
plt.scatter()
plt.show()
```

```python
Booster Shot 
```

```python
df_grouped = df_filtered.groupby('WHO_REGION').sum()

# Plotting the country vs total vaccinations
plt.figure(figsize=(10, 6))
df_grouped["PERSONS_BOOSTER_ADD_DOSE_PER100"].plot(kind='bar')

# Adding title and labels
plt.title("Persons vaccinated by the booster shot(per 100)vs WHO Region")
plt.xlabel('WHO Region')
plt.ylabel("Persons vaccinated by the booster shot(per 100)")

# Displaying the plot
plt.scatter()
plt.show()
```
